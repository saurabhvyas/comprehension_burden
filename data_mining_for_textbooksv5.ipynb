```python
import requests # for sending HTTP requests 
import pprint   # for printing stuff in a more visual way better than print

pp = pprint.PrettyPrinter(indent=4)

def concept_graph(concepts):
    # takes as an input extracted concepts of a specific section and outputs concept graph 
    
    # step 1 , determine which out of all the input concepts map directly to wikipedia articles ( call this set v)
      
        url ='https://en.wikipedia.org/w/api.php?action=query&format=json'
        
        url3 = 'https://en.wikipedia.org/w/api.php?action=query&prop=links&redirects&pllimit=500&format=json'
        #important : wikipedia api has 500 limit need to use continue option
        
        v=[]
        w={}
        concepts_relation = {}
        
        for concept in concepts:
            
            
            url2 = url + '&titles=' + concept
            r = requests.post(url2 )
            
            result_json = r.json()
            
            missing = result_json['query']['pages'].values()[0].has_key("missing")
            concept_wikiname=result_json['query']['pages'].values()[0]['title']
            pp.pprint(concept_wikiname )
            
            if missing == False:
                v.append(concept_wikiname)    
            # import instead of adding raw input concept add that concept which successfully maps and redirects to wiki
            
            url2=url
            
           # pp.pprint(v)

    # step 2 , for all concepts in V , create a dict. of all the links in these article pages , lets call this W
   
       
        for item in v:
            
            url4 = url3 + '&titles=' + item
            
            while True:
                  
                r = requests.post(url4 )
            
                result_json = r.json()
            
                if  'continue' in result_json:
                    
                    print "has more data"
                    url4=url4 + '&continue&plcontinue=' +  result_json['continue']['plcontinue']
                    #data.append(result_json['query']['pages'].values()[0]['links'])
                    #print (data)
                    w.setdefault(item,[]).extend(result_json['query']['pages'].values()[0]['links'])
                
                
                else :
                
                    url4=url4
                    print "no more data"
                   # data.append(result_json['query']['pages'].values()[0]['links'])
                    w.setdefault(item,[]).extend(result_json['query']['pages'].values()[0]['links'])
                    pp.pprint (w)
                    break
                    pp.pprint('cool'  )

           # print(result_json['query']['pages'].values()[0]['links'])
           # w.extend(result_json['query']['pages'].values()[0]['links'])
      #      w[item]=result_json['query']['pages'].values()[0]['links']
           # print("/n line break")
    #    print( w )
    
    # step 3 ,for each concept scan the  entire list W , to see if any other link corresponds to page represented by than concept
    # if it does create an edge list E and add (X,Y) to that list
    
    
        
        
        for concept in v:
            
          #  print concept
        
            for item in w:
            
             for list_item in w[item]:
                    
           #     print type(list_item)
                lower_item = list_item['title']
                lower_item=lower_item.lower()
                if concept.lower()==lower_item: # lower_item is the link of a concept list in lowercase
                    
              #      print ("hit222")
                    #concepts_relation[concept].append(item)
              #      print(concept)
              #      print(item)
                    concepts_relation.setdefault(concept,[]).append(item)
            
        print(concepts_relation)
            
       
        # concepts_relation is the adjacency list
    
    # step 4 return V and E
    
    # to do , remove self directed edges and return edge count of total graph
        
        for concept in v:
            
            for item in concepts_relation:
                
                if item==concept:
                    
                    print (item)
                    
                    for list_item in concepts_relation[item]:
                        
                        lower_item = list_item.lower()
                        
                        if lower_item == concept:
                             concepts_relation[item].remove(list_item)
                            
            print concepts_relation                
            
    
    
```
